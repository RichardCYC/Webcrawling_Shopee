```python
import requests
import pandas as pd
import jieba

headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_14_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/76.0.3809.132 Safari/537.36'}
```

```python
#----抓出搜索結果的標題----

keyword = '便盆'   #自定義搜索關鍵字
original_nameL = []

for i in range(0, 100):
    url = "https://shopee.tw/api/v2/search_items/?by=relevancy&keyword={}&limit=50&newest={}&order=desc&page_type=search".format(keyword, i*50)
    res= requests.get(url, headers = headers).json() #把結果化成json格式
    items = res.get('items')
    if len(items) != 0: #處理頁數不夠的問題
        for title in items:
            original_nameL.append(title.get('name'))
        print('完成第'+str(i+1)+'次爬蟲')
    else:
        print('newest最多到'+str(i*50)+'就沒了')
        break
```

```python
#-----透過斷詞，找出標題關鍵字-----

nameL = []

for name in original_nameL:  #做商品標題斷詞
    seg_list = jieba.lcut(name, cut_all=False)
    nameL += seg_list

stopWords=[]
remainderWords=[]

with open('/Users/richard/Documents/0Python/WebCrawling/stopWords.txt', 'r', encoding='UTF-8') as file: #開啟廢詞庫
    for data in file.readlines():
        data = data.strip()
        stopWords.append(data) 
        
remainderWords = list(filter(lambda a: a not in stopWords and a != '\n', nameL)) #用斷詞後結果與廢詞比對，留下非廢詞

dic = {}

for i in remainderWords:  #計算關鍵字出現次數
    if i.title() in dic:
        dic[i.title()] += 1
    else:
        dic[i.title()] = 1
        

df = pd.DataFrame(list(dic.items()), columns=['關鍵字', '出現次數'])

df_sort = df.sort_values(by=['出現次數'], ascending = False).reset_index(drop = True)
df_sort['搜尋結果'] = 'https://shopee.tw/search?keyword='+keyword+'%20'+df_sort['關鍵字'] #新增搜尋結果的網址，方便搜尋
```
